```python
import numpy as np
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from functions import *
target_names =  ['ACE lung','ACE seminal','ACE heart'] 
color = ['red', 'green', 'blue','orange','brown','pink','magenta']


```

### Data loading & preprocessing

```python
data, x_axis = pd.read_csv('data\\data.csv'), np.load('data\\x_axis.npy')
```

```python
Smooth, Standartization, Baselin_corr, median_filter_parameter, first_diff, cut_range_from = True, True, True, False, False, 170

data = preprocess_data(data, Smooth, Standartization, Baselin_corr, median_filter_parameter, first_diff, cut_range_from)[0] 
x_axis = np.array(x_axis[cut_range_from:])
```

### Training model


```python
X_train, X_test, y_train, y_test = train_test_split(np.array(data)[:,:-2],
                                                    np.array(data)[:,-2:], test_size = 0.5)
X_train,y_train = add_noise(X_train,y_train)
lda = LDA(n_components = 2, store_covariance = True)
train_model(X_train,y_train, lda)
```

### Testing quality

```python
y_pred = lda.predict(X_test)
classification_rep(y_test,y_pred)
plot_2d_lda_space(lda,X_test,y_test)
# plot_confusion_matrix(y_test, y_pred)

```
